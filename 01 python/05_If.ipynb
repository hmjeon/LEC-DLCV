### **05. 조건문(if)**

#### **if, elif, else**
- 모든 조건은 boolean으로 표현

```python
if 6 >= 5:
    print('6 is greater than 5')
    print('This is true')
    
print('This code is not belongs to if statements')
```

```python
if 6 == 5:
    print('6 is greater than 5')
    print('This is true')
    
print('This code is not belongs to if statements')
```

#### **논리연산자(and, or, not)** 
- 비교연산자(>. <. <=, <=, ==, !=)와 함께 사용
- 논리연산자 우선순위: not > and > or
- 논리표

```python
a = 1; b = 2; c = 3
if (a == 1 or b == 3) and c == 3:
    print('This is true')
```

```python
if not a == 10:
    print('a is 1')
```

#### **False로 간주되는 값**
- None, 0, 0.0, '', [], (), {}, set()
- 그외는 모두 True 값

```python
a = []
if not a:
    print('print')
```

#### **if, elif, else**

```python
# 홀수 및 짝수 구별
a = 10
if a % 2 == 0:     # 짝수인지 판별
    print('even')
else:
    print('odd')
```

```python
a = 16
if a % 4 == 0:
    print('a can be divided by 4')
elif a % 4 == 1:
    print('a % 4 is 1')
elif a % 4 == 2:
    print('a % 4 is 2')
else:
    print('a % 4 is 3')
```

```python
a = 16
if a % 4 == 0:
    print('a can be divided by 4')
if a % 4 == 1:
    print('a % 4 is 1')
if a % 4 == 2:
    print('a % 4 is 2')
else:
    print('a % 4 is 3')
```

#### **중첩 조건문(Nested Condition)**

```python
a = 1; b = 2; c = 3

if a == 1:
    if c == 3:
        if b == 2:
            print('a is 1 and b is 2')
        else:
            print('a is 1 and b is not 2')
```
