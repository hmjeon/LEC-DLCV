# **모듈(Module)**

## import 키워드

```python
import math

print(math.pi)
print(math.cos(100))
```

## from import 키워드
- 해당 모듈에서 특정한 타입만 import
- import *: 해당 모듈내에 정의된 모든 함수 사용가능 (권장 X)

```python
from math import pi
from math import cos

print(round(pi, 4))
print(cos(100))
```

## as 키워드

```python
import math as m
print(m.exp(3))
print(m.cos(100))
```
