### **09. Lambda 함수**

#### **Lambda 작성성**
- 한줄의 간단한 함수

```python
def square(x):
    print(x**2)
square(5)

square = lambda x:print(x**2)
square(5)
```

```python
def add(x, y):
    return x + y

add_lamda = lambda x, y:x + y

add1 = add(10, 20)
add2 = add_lamda(10, 20)

print(add1, add2)
```

```python
def str_len(s):
    return len(s)

str_len('goods')
```

```python
strings = ['Mark', 'Hyungmin', 'Tyson', 'Sakul']
#strings.sort(key=str_len)
strings.sort(key=lambda s:len(s))

print(strings)
```

#### **Lamda 함수 사용 예**
- filter 및 map

```python
# filter: 특정 조건을 만족하는 요소만 남기고 필터링
nums = [1, 2, 3, 6, 8, 9, 10, 11, 13, 15]
list(filter(lambda n:n%2==0, nums))
```

```python
# map: 각 원소를 주어진 수식에 따라 변형하여 새로운 리스트를 반환
nums = [1, 2, 3, 6, 8, 9, 10, 11, 13, 15]
list(map(lambda n:n**2, nums))
```

```python
nums = [1, 2, 3, 6, 8, 9, 10, 11, 13, 15]
list(map(lambda n:n%2==0, nums))
```
