# **문자열(String)**

```python
a = 'Hello World'
b = "Hello' World"
c = '''Hello World.
It's wonderful world'''

print(a)
print(b)
print(c)
```

## 이스케이프 문자(Escape String)
- `\n`: 개행, `\t` : tab

```python
print('Hello \nWorld\n\n')
print('Deep\Learning')
```

## 문자열 인덱싱(Indexing)

```python
a = 'Hello World'    # 공백을 포함하여 총 11개의 문자로 구성
print(a[10])
```

## -1 인덱스
- -1은 가장 마지막 인덱스를 의미

```python
a = 'Hello World'

print(a[0])
print(a[10])

print(a[-1])
print(a[-11])

# print(a[11])
# print(a[-12])
```

## 문자열 슬라이싱(Slicing)

```python
a = 'Hello world'

print(a[0:11])
print(a[0:1])

print(a[:5])
print(a[3:])

print(a[:])
```

## 문자열 멤버함수
- 변수 다음에 '.'을 사용하여 멤버함수를 호출
- upper(), lower(), replace(), format(), split() 등

```python
a = 'hello world'
a.upper() #.lower()
```

```python
a = 'hello world'
a.replace('l', 'j')
```

```python
height = 183.5
weight = 85.5

a = '키는 {}cm 이고 몸무게는 {}kg 입니다.'.format(height, weight)
print(a)
```

```python
a = 'This is Hyungmin Jun.'
a.split(' ')
```
