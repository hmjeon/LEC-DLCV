### **08. 함수**

#### **내장함수(Built-in Function)**

```python
a = [1, 2, 3, 4]
length = len(a)
print(length)

_sum = sum(a)
print(_sum)
```

#### **사용자 정의 함수(User-defined Function)**

```python
def add(a, b):
    n = a + b
    return n

c = add(1, 2)
print(c)
```

#### **함수 네이밍(Naming)**

```python
def substract(x, y):
    sub = x - y
    return sub

print(substract(4, 3))
```

#### **함수 인자(Parameter, Argument) 및 함수 내부 값 전달(Return)**
- 기본인자(Default Parameter)
- print 함수의 기본인자
- 함수정의 시, 기본인자 뒤에 일반인자 사용불가

```python
def add(x, y=4, z=5):
    a = x + y + z
    return a

print(add(10))
print(add(10, 1, 2))
```

```python
print(1, 2, 3, sep='!', end='%%%')
print()
print(2, 3, 4, sep='p')
```

```python
def add(x, y, z):
    a = x - y + z
    return a

print(add(1, 2, 3))
print(add(x=1, z=3, y=2))
```

#### **함수 내부의 복수 값 전달(Multiple Return)**

```python
def add_mul(x, y):
    s = x + y
    m = x * y

    return s, m

a = add_mul(20, 3)    # Return 값은 튜플로 전달
print(a)

b,c = add_mul(20, 3)  # 각각 변수에 값을 전달
print(b,c)
```

#### **변수범위(Variable Scope)**

```python
num1 = 10
num2 = 30

def test(num1, num2):
    print(num1, num2)
    return num1 + num2

num3 = test(30, 40)
print(num1, num2)
```

#### **가변길이 인자(Variable Length Argument)**
- 전달되는 파라미터의 개수가 고정적이지 않은 경우

> ***args**    : 파라미터를 튜플의 형태로 전달

> ****kwargs** : 파리미터를 딕셔너리 형태로 전달(키워드 파라미터)

```python
# Variable length arguments
def test(*args):
    for item in args:
        print(item)
    
test(10, 30, 40, 50, 60, 70)

# Keyword arguments
def test2(**kwargs): 
    for key, value in kwargs.items():
        print('key:', key, ', value:', value)

test2(a=1, b=2)
```
