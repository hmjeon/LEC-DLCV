#### **ndarray를 활용한 2차원 벡터 곱셈 연산 함수 만들기**

```python
# 임의 크기의 2차원 배열 a, b를 생성
# a[a1][a2], b[b1][b2]에서 a2과 b1은 항상 동일한 크기
# a*b 매트릭스 곱셉을 했을 때, c[a1][b2]가 얻어짐
# a*b 매트릭스 곱셉에 대한 사용자 함수 정의

np.random.seed(10)    # Seed 함수, 동일한 Random 수 생성

# 0, 1사이의 분포로 랜덤한 ndarray 생성
a = np.random.rand(4, 5)
b = np.random.rand(5, 3)

print('a = \n', a)
print('b = \n', b)

def mul_mat(a, b):
  c = np.zeros((len(a), len(b[0])))
  for i in range(len(a)):
    for j in range(len(b[0])):
      temp = 0
      for k in range(len(a[0])):
        temp = temp + a[i][k] * b[k][j]
      c[i][j] = c[i][j] + temp
  return c

c = mul_mat(a, b)
print('c = \n', c)
print('c = \n' ,a.dot(b))
```

#### **Indexing and Slicing**

```python
import numpy as np

# 1차원 배열 인덱싱
x = np.arange(10)
print(x)

x[3] = 100
print(x)

# 2차원 배열 인덱싱
x = np.arange(10).reshape(2, 5)
print(x)

print(x[1,-1])
print(x[0])
```

```python
x = np.arange(10)
print(x)
print(x[1:])

x = np.arange(10).reshape(2, 5)
print(x)
print(x[0, :2])
print(x[:2, :2])
```
